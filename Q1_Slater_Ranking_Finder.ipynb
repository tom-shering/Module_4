```python
from itertools import permutations 

# Currently this is very static. 

L1 = [0, 0, 0, 0, 1]
L2 = [1, 0, 0, 0, 1]
L3 = [1, 1, 0, 1, 0]
L4 = [1, 1, 0, 0, 1]
L5 = [0, 0, 1, 0, 0]

print("**** Majority Graph being assessed:\n")
print(L1)
print(L2)
print(L3)
print(L4)
print(L5)
print("\n**** Preference orders with a low Slater ranking:\n")


pref_order_perms = list(permutations([1, 2, 3, 4, 5]))


class Majority_Graph():

  def __init__(self):
    self.SR = 0
    self.best_SR = 10
    #self.possible_preference_orders = []
    #self.best_preference_orders = []
    self.L1i = 0
    self.L2i = 0
    self.L3i = 0
    self.L4i = 0
    self.L5i = 0
    self.L1_alt = [0, 0, 0, 0, 0]
    self.L2_alt = [0, 0, 0, 0, 0]
    self.L3_alt = [0, 0, 0, 0, 0]
    self.L4_alt = [0, 0, 0, 0, 0]
    self.L5_alt = [0, 0, 0, 0, 0]

  def find_Slater_rankings(self, po):
    # Set indexes of the locations:
    self.L1i = po.index(1)
    self.L2i = po.index(2)
    self.L3i = po.index(3)
    self.L4i = po.index(4)
    self.L5i = po.index(5)

    # Establish edges for MG:
    if self.L1i < self.L2i:
      self.L1_alt[1] = 1
    else: 
      self.L2_alt[0] = 1
    
    if self.L1i < self.L3i:
      self.L1_alt[2] = 1
    else: 
      self.L3_alt[0] = 1

    if self.L1i < self.L4i:
      self.L1_alt[3] = 1
    else: 
      self.L4_alt[0] = 1

    if self.L1i < self.L5i:
      self.L1_alt[4] = 1
    else: 
      self.L5_alt[0] = 1

    if self.L2i < self.L3i:
      self.L2_alt[2] = 1
    else: 
      self.L3_alt[1] = 1

    if self.L2i < self.L4i:
      self.L2_alt[3] = 1
    else: 
      self.L4_alt[1] = 1
    
    if self.L2i < self.L5i:
      self.L2_alt[4] = 1
    else: 
      self.L5_alt[1] = 1
    
    if self.L3i < self.L4i:
      self.L3_alt[3] = 1
    else: 
      self.L4_alt[2] = 1

    if self.L3i < self.L5i:
      self.L3_alt[4] = 1
    else: 
      self.L5_alt[2] = 1
    
    if self.L4i < self.L5i:
      self.L4_alt[4] = 1
    else: 
      self.L5_alt[3] = 1
    

    # Compare MG discrepancies:
    for i in range(5):
      if self.L1_alt[i] == 0 and L1[i] == 1:
        self.SR += 1
      if self.L2_alt[i] == 0 and L2[i] == 1:
        self.SR += 1
      if self.L3_alt[i] == 0 and L3[i] == 1:
        self.SR += 1
      if self.L4_alt[i] == 0 and L4[i] == 1:
        self.SR += 1
      if self.L5_alt[i] == 0 and L5[i] == 1:
        self.SR += 1

    
    if self.SR < 3:
      print(self.L1_alt)
      print(self.L2_alt)
      print(self.L3_alt)
      print(self.L4_alt)
      print(self.L5_alt)
      print(f"Slater ranking for {po} is {self.SR}.")
    if self.SR <= self.best_SR:
      self.best_SR = self.SR


possible_preference_orders = []
best_preference_orders = []
lowest_SR = 10

for i in range(len(pref_order_perms)):
  MG2 = Majority_Graph()
  MG2.find_Slater_rankings(pref_order_perms[i])
  preference_order_ranking = [pref_order_perms[i], MG2.SR]
  possible_preference_orders.append(preference_order_ranking)

#find lowest SR:
for j in range(len(possible_preference_orders)):
  if possible_preference_orders[j][1] < lowest_SR:
    lowest_SR = possible_preference_orders[j][1]
# Collect pref. orders with this SR:
for k in range(len(possible_preference_orders)):
  if possible_preference_orders[k][1] == lowest_SR:
      best_preference_orders.append(possible_preference_orders[k][0])

print(f"\n**** The lowest Slater ranking for Q1 is {lowest_SR}.")
print(f"**** Preference orders with this Slater ranking for Q1: {best_preference_orders}.")


L1 = [0, 0, 0, 0, 1]
L2 = [1, 0, 0, 0, 1]
L3 = [1, 1, 0, 1, 0]
L4 = [1, 1, 0, 0, 1]
L5 = [0, 0, 1, 0, 0]

```
